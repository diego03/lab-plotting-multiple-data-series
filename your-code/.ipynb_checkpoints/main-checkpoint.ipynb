# Plotting Multiple Data Series

Complete the following set of exercises to solidify your knowledge of plotting multiple data series with pandas, matplotlib, and seaborn. Part of the challenge that comes with plotting multiple data series is transforming the data into the form needed to visualize it like you want. For some of the exercises in this lab, you will need to transform the data into the form most appropriate for generating the visualization and then create the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('../data/liquor_store_sales.csv')
data.head()
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
#One way is through the groupby methond
data_type = data.groupby(['ItemType'])[['RetailSales', 'RetailTransfers','WarehouseSales']].sum()
data_type
```

```python
#Another is with pivot tables
pivot = data.pivot_table(index='ItemType', aggfunc={'RetailSales': 'sum', 'RetailTransfers':'sum', 'WarehouseSales':'sum'})
pivot
```

```python
#Creating the Graph:
pivot.plot.bar()

#Also
data_type.plot.bar()
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
#I don't entirely understand what you mean for sales mix but I'll try according to what I'm understanding
#Create new column with total sales: total sales being retail plus warehouse sales
data['TotalSales'] = data['RetailSales'] + data['WarehouseSales']

#Create Pivot Table indexing by Supplier
pivot2 = data.pivot_table(index='Supplier', aggfunc = {'RetailSales':'sum', 'WarehouseSales':'sum', 'TotalSales':'sum'})

#Order in a descending fashion (from most to least)
pivot2 = pivot2.sort_values(by = 'TotalSales', ascending = False )

#Obtain only the top 10 suppliers by totalsales
pivot2_top10 = pivot2[0:10]

#Graph by type of sales (including totalSales)
pivot2_top10.plot.barh()
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
#Create Pivot Table indexing by month and showing RetailSales, RetailTransfers and WarehouseSales
pivot_time = data.pivot_table(index=['Year','Month'], aggfunc = {'RetailTransfers':'mean', 'WarehouseSales':'mean', 'RetailSales':'mean'})

#Order in an ascending fashion (from most to least)
pivot_time = pivot_time.sort_values(by = ['Year','Month'], ascending = [True,True] )

#Graph by month overtime
pivot_time.plot.line(figsize=(15,7))
```

## 4. Plot the same information as above but as a bar chart.

```python
pivot_time.plot.bar(figsize=(15,7))
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
#Pivot Table by ItemType
pivot3 = data.pivot_table(index = ['Year','Month'], columns=['ItemType'], aggfunc = {'RetailSales':'sum'})

#Create Graph
pivot3.plot.line(figsize=(15,7))
```

## 6. Plot the same information as above but as a bar chart.

```python
#Create Graph
pivot3.plot.bar(figsize=(15,7))
```

## 7. Create a scatter plot showing the relationship between Retail Sales (x-axis) and Retail Transfers (y-axis) with the plot points color-coded according to their Item Type.

*Hint: Seaborn's lmplot is the easiest way to generate the scatter plot.*

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='RetailSales', y='RetailTransfers', data=data)
```
